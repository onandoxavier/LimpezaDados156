```python
import pandas as pd
```

```python
tabela = 'public.a1656155_156'
colunas = "SOLICITACAO, TIPO, ORGAO, DATA_ASS, HORARIO, ASSUNTO, SUBDIVISAO, DESCRICAO, LOGRADOURO_ASS, CEP_ASS, NUMERO_ASS, BAIRRO_ASS, REGIONAL_ASS, MEIO_RESPOSTA, OBSERVACAO, SEXO, BAIRRO_CIDADAO, REGIONAL_CIDADAO, DATA_NASC, TIPO_CIDADAO, ORGAO_RESP, RESPOSTA_FINAL, GEOM"

df = pd.read_csv('Resultado/inundacao156latlng.csv', sep=';')
df.info()
```

```python
# Retira tuplas duplicadas
df.drop_duplicates(subset ="SOLICITACAO", keep = False, inplace = True)
df.info()
```

```python
# Criando e escrevendo em arquivos de texto (modo 'w').
arquivo = open('Resultado/insertSql.txt','w')

for index, row in df.iterrows():
    latlng = row['GEOM'].split(',')
    lat = latlng[0][1:]
    lng = latlng[1][1:-1]
    
    geom = f'ST_SetSRID(st_MakePoint(cast({lng} as double precision),cast({lat}  as double precision)),29193)'
        
    # Como o campo 'Resposta_final' não era essencial para o fim da pesquisa e estava limitado ao tamanho de 200
    # caracteres no banco eu gerei uma substring que não gerasse erro ao ser inserida no banco
    if(len(row['RESPOSTA_FINAL']) >= 200):
        respostaFinal = row['RESPOSTA_FINAL'][0:198].strip()
    else:
        respostaFinal = row['RESPOSTA_FINAL']    
    
    # Adiciona o valor vazio ao campo ao inves de 'nan'
    if(pd.isna(row['SEXO'])):
        sexo = ''
    else:
        sexo = row['SEXO']
        
    # Adiciona o valor NULO aos campos não preenchidos
    if(pd.isna(row['DATA_ASS']) or row['DATA_ASS'] == ''):
        data_ass_val = 'NULL'
    else:
        data_ass_val = f"\'{row['DATA_ASS']}\'"
        
    # Adiciona o valor NULO aos campos não preenchidos    
    if(pd.isna(row['DATA_NASC']) or row['DATA_NASC'] == ''):
        data_nasc_val = 'NULL'
    else:
        data_nasc_val = f"\'{row['DATA_NASC']}\'"   
    
    values = f"\'{row['SOLICITACAO']}\', \'{row['TIPO']}\', \'{row['ORGAO']}\', {data_ass_val}, \'{row['HORARIO']}\', \'{row['ASSUNTO']}\', \'{row['SUBDIVISAO']}\', \'{row['DESCRICAO']}\', \'{row['LOGRADOURO_ASS']}\', \'{row['CEP_ASS']}\', \'{row['NUMERO_ASS']}\', \'{row['BAIRRO_ASS']}\', \'{row['REGIONAL_ASS']}\', \'{row['MEIO_RESPOSTA']}\', \'{row['OBSERVACAO']}\', \'{sexo}\', \'{row['BAIRRO_CIDADAO']}\', \'{row['REGIONAL_CIDADAO']}\', {data_nasc_val}, \'{row['TIPO_CIDADAO']}\', \'{row['ORGAO_RESP']}\', \'{respostaFinal}\', {geom};"
    
    comando = f"INSERT INTO {tabela} ({colunas}) VALUES ({values})\n"    
    arquivo.write(f"{comando}")
arquivo.close()
```
